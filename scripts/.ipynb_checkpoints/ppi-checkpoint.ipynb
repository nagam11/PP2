```python
import argparse
import os
import numpy as np
import biovec
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import StratifiedKFold
```

## Create input vectors

```python
vectors_file = "../output/sp_words-n3-w5-s100-k10-1.w2v" 
ppi_file = "../data/ppi_data.fasta"
os.path.isfile(vectors_file), os.path.isfile(ppi_file)
```

```python
names = []
seqs = []
ppis = []
# read data
with open(ppi_file) as file:
    i = 0
    for line in file:
        if (i % 3 == 0):
            # name
            names.append(line[1:].strip())
        elif (i % 3 == 1):
            # sequence
            seqs.append(line.strip())
        elif (i % 3 == 2):
            # labels
            ppis.append(line.strip())
        i = i + 1
len(names)
```

```python
vectors = {}
with open(vectors_file) as file:
    i = 0
    for line in file:
        aa_vec = line.split()
        feature = np.array([float(x) for x in aa_vec[1:]])
        vectors.update({aa_vec[0]: feature})
        i = i + 1
len(vectors)
```

```python
def convert_long_seq_gram(long_gram, offset=3):
    seq_gram_strings = [long_gram[j:j+offset] for j in range(2*offset-1)] # get short grams within long gram
    seq_gram_vectors = [vectors[gram] for gram in seq_gram_strings] # convert into feature vectors
    gram_sum = np.array(seq_gram_vectors).sum(axis=0) # sum up short grams
    return gram_sum

def convert_long_ppi_gram(long_gram, offset=3):
    #ppi_gram_strings = [long_gram[j:j+offset] for j in range(2*offset-1)] # get short grams
    #ppi_gram_vectors = [translate_ppi(gram) for gram in ppi_gram_strings] # convert into feature vectors
    #gram_sum = np.array(ppi_gram_vectors).sum(axis=0) # sum up short grams
    #gram_sum = np.divide(gram_sum, len(ppi_gram_vectors)) # normalise over number of grams
    #return np.round(gram_sum)
    ppi_vector = translate_ppi(long_gram)
    labels = {x:list(ppi_vector).count(x) for x in set(ppi_vector)}
    return max(labels, key=labels.get)

def translate_ppi(ppi_string):
    return np.array(list(map(lambda x:0 if x=="-" else 1, ppi_string)))
```

```python
offset = 3
features = []
labels = []
for i in range(offset,len(names)-offset+1):
    features.append(convert_long_seq_gram(seqs[i])) # sequence residues
    labels.append(convert_long_ppi_gram(ppis[i])) # ppi labels per residue
len(features), len(labels)
```

## Train Model

```python
X = np.array(features)
y = labels
```

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(25, 50, 100),  
    learning_rate_init=0.001,
    n_iter_no_change=100,
    random_state=42)
```

```python
mlp.fit(X, y)
```

```python
StratifiedKFold(n_splits=2, random_state=None, shuffle=False)
skf = StratifiedKFold(n_splits=10, random_state=42)

for train_index, test_index in skf.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    #X_train, X_test = X[train_index], X[test_index]
    #y_train, y_test = y[train_index], y[test_index]
```
